```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import scipy

# divide the data between trials
```

```python
# Convertes voltage to Force in grams
weights = [0, 120, 150, 200, 250, 300, 350, 400, 450, 500, 550, 600, 650, 700, 750, 800, 900, 1000]
value = [0, 140, 300, 580, 640, 740, 770, 800, 870, 880, 890, 900, 905, 915, 927, 925, 935, 945]
normalized_value = [x / 1023 * 5 for x in value]

def get_force (yvalues):
    x = weights
    y = normalized_value
    xinterp = np.interp(yvalues, y, x)
    xinterp = (xinterp/1000)*9.8
    
    return xinterp
```

```python
# Convertes analogue values to voltage
def arduino_to_voltage(arduino_values):
    voltage_FSR = (arduino_values / 1023)*5
    return voltage_FSR
```

```python
# Extract 1st and 2nd trial of all participants
experiments = {}
for person in range(1, 12):
    for trial in (1, 2):
        name = f"Data/arduino_person_{person:03d}_exp_2_trial_{trial}.csv"
        category = (person, trial % 2 + 1) if 11 > person > 5 else (person, trial)
        columns = ["position", "force"]
        if person > 4:
            columns = ["trigger"] + columns
        experiments[category] = pd.read_csv(name, encoding="latin", header=None, names=columns)
```

```python
for category, data in experiments.items():
    plt.figure(figsize=(20,5))
    plt.scatter(data.index, data.loc[:, "force"])
    plt.title(f"Person {category[0]} Trial {category[1]}")
    plt.show()
## calculate frequency and change x axis to time 
```

```python
# Finding each trial within the data
l = []
force_thresholds = {
    (4, 1): 600,
    (4, 2): 600,
    (5, 1): 700,
}

all_subsets = []
all_subsets_ = []
for category, data in experiments.items():
    print(f"Person {category[0]} Trial {category[1]}")
    subset = data[data["force"] >= force_thresholds.get(category, 800)].reset_index().copy()
    subset["group_change"] = (subset["index"] - subset["index"].shift(1) - 1) != 0
    subset["group_n"] = subset["group_change"].cumsum()
    subset = subset.groupby("group_n")["index"].agg(["min", "max", "count"])
    subset = subset[subset["count"] > 200]
    
    print(f"\tFound {subset.shape[0]} groups")
 
    for group, row in subset.iterrows():
        group_subset = data.loc[row["min"]+20:row["max"]-20].copy()

        #Uncomment below to plot each result
        plt.scatter(group_subset.index, group_subset["force"])
        plt.title(f"group {group} from {row['min']} to {row['max']}")
        plt.show()
        l.append((category[0], category[1], row['min']+20, row['max']-20))
        
        group_subset["person"] = category[0]
        group_subset["trial"] = category[1]
        all_subsets.append(group_subset)
        all_subsets_.append(group_subset)
        print(f"\tSegment average: {group_subset['force'].mean():.2f}. Std: {group_subset['force'].std():.2f}")

    print()

all_subsets = pd.concat(all_subsets)
print(f"Overall: {all_subsets['force'].mean()}, std: {all_subsets['force'].std()}")

all_subsets.groupby("trial")["force"].agg(["mean", "std"])

## separe each trial of each subject
```

```python
for index, value in enumerate(l):
    print(index, value)
    
groups_to_delete= [5, 11, 22, 23, 29, 35, 36, 45, 49, 74, 75, 76, 77, 79, 85, 87, 96, 103, 107, 112, 120, 121, 127, 130]
final_groups = []

for index, value in enumerate(l):
    if index not in  groups_to_delete:
        final_groups.append(value)
```

```python
all_subsets["force"] = arduino_to_voltage(all_subsets["force"])
all_subsets["force"] = get_force(all_subsets["force"])
```

```python
task2_results = []
trial = 0
previous_subtask = 1
for person, subtask, start_idx, end_idx in final_groups:
    if subtask != previous_subtask:
        trial = 1
        previous_subtask = subtask
    else:
        trial += 1
        
    subset = all_subsets[(all_subsets["person"] == person) & (all_subsets["trial"] == subtask)]
    subset = subset.loc[start_idx:end_idx]
    subset.to_csv(f"task2_results/subject_{person}_subtask_{subtask}_trial{trial}.csv", index=False, columns=["force"])
    task2_results.append(subset)
task2_results = pd.concat(task2_results)
```

```python
task2_results
```

```python
#Box plot
without_feedback = task2_results.loc[task2_results["trial"] == 1, "force"]
with_feedback = task2_results.loc[task2_results["trial"] == 2, "force"]

plt.figure(figsize=(5, 5))
plt.boxplot([without_feedback, with_feedback], labels=['without feedback', 'with feedback'])

plt.title('Force applied', fontsize="18")
plt.ylabel('Force (N)', fontsize="16")
plt.xticks(fontsize=16) 
plt.axhspan(3.92, 5.88, facecolor='blue', alpha=0.2)
plt.show()
```

```python
#Histogram without feedback
sns.histplot(without_feedback, bins=20)
plt.title('Histogram of force without feedback', fontsize="18")
plt.xlabel('Force (N)', fontsize="16")
plt.ylabel('Count', fontsize="16")
plt.axvspan(3.92, 5.88, facecolor='blue', alpha=0.2)
plt.show()
```

```python
#Histogram with feedback
sns.histplot(with_feedback, bins=20, color='green')
plt.title('Histogram of force with feedback', fontsize="18")
plt.xlabel('Force (N)', fontsize="16")
plt.ylabel('Count', fontsize="16")
plt.axvspan(3.92, 5.88, facecolor='blue', alpha=0.2)
plt.show()
```

# Analysis of results

```python
lw = len(with_feedback)
lw_in = ((with_feedback > 3.92) & (with_feedback < 5.88)).sum()
force_range_with = lw_in/lw

print(f" Amount of time between the range with feedback: {force_range_with}")
```

```python
lwo = len(without_feedback)
lwo_in = ((without_feedback > 3.92) & (without_feedback < 5.88)).sum()
lll = (without_feedback < 5.88).sum()
force_range_without = lwo_in/lwo
above = 1 - lll/lwo

print(f" Amount of time between the range without feedback: {force_range_without}")
print(f" Amount of time above the range without feedback: {above}")
```

```python
with_feedback_iqr = with_feedback.quantile(0.75) - with_feedback.quantile(0.25)
without_feedback_iqr = without_feedback.quantile(0.75) - without_feedback.quantile(0.25)
difference = without_feedback_iqr / with_feedback_iqr

print(f" Difference of inter-quartile range: {difference}")
```
